```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('C:/Python/Python39/Scripts/PythonProjects/7 Finance-Data/DataSet/Finance_data.csv')
print(f'Count attributes of dataset: {len(data.columns)}')
print(f'Count rows of dataset: {len(data)}')
```

```python
plt.figure(1)
plt.figure(figsize=(7,5))

sns.countplot(data['gender'], palette="Set2", edgecolor='black', linewidth=1)
plt.ylabel='Count of People'
plt.xlabel='Gender'

plt.show()
```

```python
data.groupby('gender').age.plot(kind='kde')
plt.xlabel('age')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data['age'], palette="Set3", linewidth=1, edgecolor='black')
plt.title("Count of investers by age")
plt.show()

plt.figure(figsize=(7,5))
sns.countplot(data['age'], hue=data['gender'], linewidth=1, edgecolor='black')
plt.title("Count of investers by age")
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1, 2, 1)
sns.countplot(x=data['gender'], hue=data['Investment_Avenues'], palette='summer', linewidth=1.2, edgecolor='black')
plt.title('INVESTMENT AVENUE')

plt.subplot(1, 2, 2)
sns.countplot(x=data['gender'], hue=data['Stock_Marktet'], palette='hot', linewidth=1.2, edgecolor='black')
plt.title('STOCK MARKET')

plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=data['Factor'], palette='coolwarm', linewidth=1, edgecolor='black')
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=data['gender'], hue=data['Factor'], palette='Oranges', linewidth=1, edgecolor='black')
plt.title('Gender')
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1, 2, 1)
sns.countplot(x=data['Duration'], hue=data['gender'], palette='viridis', linewidth=1, edgecolor='black')
plt.title('DURATION')

plt.subplot(1, 2, 2)
sns.countplot(x=data['Invest_Monitor'], hue=data['gender'], palette='seismic', linewidth=1, edgecolor='black')
plt.title('INVESTMENT MONITORING')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=data['What are your savings objectives?'], hue=data['gender'],palette='Oranges',linewidth=1,edgecolor='black')
plt.title('Gender')
plt.show()
```

```python
sns.jointplot(data=data, x="age", y="Equity_Market", kind="kde", color="y", height=7)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))

sns.countplot(ax=axes[0], x=data['Duration'], edgecolor='black', linewidth=1)
axes[0].set_title('Prefer to keep your money')

sns.countplot(ax=axes[1], x=data['Invest_Monitor'], edgecolor='black', linewidth=1)
axes[1].set_title('Invest Monitor')
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x=data['Factor'], palette='Oranges', linewidth=1, edgecolor='black')
plt.title('Factor of Investment')
plt.show()
```
